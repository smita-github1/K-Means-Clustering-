 OBJECTIVE:

 To Divide countries into Clusters based on their Longitude and Latitude.

 Dataset:

 Rows-6
 Columns-4 Labels are-

 1.country ( Datatype- Categorical)
 2.Latitude( Datatype- Numerical)
 3.Longitude(Datatype- Numerical)
 4.Language (Categorical)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/3.01. Country clusters ss.csv")
data
# in this data we don't have any dependent and independent variables
# we have only input data,so we will cluster
```

```python
data.shape
```

```python
# Filtering Labels we need
new_data=data.iloc[:,1:3]
new_data.head()
```

```python
#importing the algorithm
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=3)      # k=3 ,observing data shows 3 groups
```

```python
# we will train the model
Kmeans.fit(new_data)
```

```python
# clusters are on the basis of longitude and latitude
clusters= Kmeans.fit_predict(new_data)
```

```python
print(clusters)
```
